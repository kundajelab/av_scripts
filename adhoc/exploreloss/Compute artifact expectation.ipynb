```python
import numpy as np

bins = np.array([x*1e-5 for x in range(2*(10**5))])
density_in_bin = (np.log(10)*(np.exp(-np.log(10)*bins+(0.5*1e-5)) - 
                   np.exp(-np.log(10)*bins-(0.5*1e-5))))
stdev_of_bins = np.sqrt(3e9*density_in_bin*(1-density_in_bin))
```

```python
%matplotlib inline
import matplotlib
from matplotlib import pyplot as plt
print(np.max(density_in_bin*3e9 + 6*stdev_of_bins))
plt.plot(bins, density_in_bin*3e9 + 6*stdev_of_bins)
plt.show()
plt.plot(bins, stdev_of_bins)
plt.show()
```
