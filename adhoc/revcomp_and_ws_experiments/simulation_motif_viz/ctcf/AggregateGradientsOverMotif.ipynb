```python
from __future__ import division
from __future__ import print_function
from collections import OrderedDict
%matplotlib inline

from simdna.synthetic import read_simdata_file
data = read_simdata_file("DensityEmbedding_prefix-ctcf_motifs-CTCF_known1_min-1_max-2_mean-2_zeroProb-0_seqLength-400_numSeqs-1000.simdata.gz")
```

```python
from avutils.util import seq_to_one_hot
onehot_encoded_inp = [seq_to_one_hot(x) for x in data.sequences]
```

```python
#model_prefix = "record_4_model_4DFtL" #revcomp
#model_prefix = "record_3_model_1LMuB" #siamese
model_prefix = "record_1_model_xNgm1" #standard model

model_weights = "model_files/"+model_prefix+"_modelWeights.h5"
model_json = "model_files/"+model_prefix+"_modelJson.json"
```

```python
from keras.models import model_from_json
model = model_from_json(open(model_json).read())
model.load_weights(model_weights)
```

```python
import numpy as np
rand_inp = np.random.random((10,400,4))
#verify rc property
print(np.max(np.abs(model.predict(rand_inp) - model.predict(rand_inp[:,::-1,::-1]))))
```

```python
#compile the gradient computation function
import theano
from theano import tensor as T
from keras import backend as K

grad_calc_func = theano.function(
[model.input_layers[0].input], T.grad(K.sum(model.layers_by_depth[1][0].output[:,0]),
                                model.input_layers[0].input))
```

```python
from deeplift.util import run_function_in_batches

grads = np.array(run_function_in_batches(func=grad_calc_func,
                        input_data_list=[onehot_encoded_inp],
                        learning_phase=None,
                        batch_size=50,
                        progress_update=1000))
```

```python
#grad_times_inp = grads*(onehot_encoded_inp - np.array([0.3, 0.2, 0.2, 0.3])[None,None,:])
```

```python
mean_imp_on_embedding = None
num_embeddings = 0
for embeddings_for_sequence, imp_for_seq in zip(data.embeddings, grads):
    for embedding in embeddings_for_sequence:
        num_embeddings += 1
        startPos = embedding.startPos
        imp_scores_on_embedding = imp_for_seq[startPos:(startPos+len(embedding.what.string)),:]
        imp_scores_on_embedding = imp_scores_on_embedding - np.mean(imp_scores_on_embedding,axis=1)[:,None]
        if ('revComp' in embedding.what.stringDescription):
            imp_scores_on_embedding = imp_scores_on_embedding[::-1,::-1]
        if (mean_imp_on_embedding is None):
            mean_imp_on_embedding = imp_scores_on_embedding
        else:
            mean_imp_on_embedding = mean_imp_on_embedding + imp_scores_on_embedding
mean_imp_on_embedding = mean_imp_on_embedding/float(num_embeddings)
```

```python
from deeplift.visualization import viz_sequence
viz_sequence.plot_weights(mean_imp_on_embedding, figsize=(10,2))
```
